Получить список поддиректорий в директории Mdl и csv файлов

```python
import os
import pandas as pd

# Replace 'your_directory_path' with the path of the directory you want to work with
directory_path = './Mdl/'

# Initialize an empty list to store full file paths of CSV files
csv_file_paths = []

# Iterate through the items (files and folders) in the directory
for root, dirs, files in os.walk(directory_path):
    for file in files:
        if file.endswith('.csv'):
            csv_file_paths.append((os.path.join(root, file).replace('\\', '/')))

# Create a DataFrame using pandas
data = {'CSV File Paths': csv_file_paths}
df = pd.DataFrame(data)

print(df)
```

Получить все CSV файлы в папке

```python
df_err = pd.DataFrame(columns = ['CSV File Paths', 'Max_err', 'Max_sqr_err', 'MSE', 'MSQRT'])
for file in df["CSV File Paths"]:
    print(file)
    df_1 = pd.read_csv(file,sep=',',encoding='utf-8')
       
    max_err = df_1['Res_hat'].abs().max()
    max_sqr_err = df_1['Res_hat'].pow(2).max()
    mse = df_1['Res_hat'].pow(2).mean()
    msqrt = np.sqrt(mse)
    new_row ={'CSV File Paths': file, 'Max_err': max_err, 'Max_sqr_err': max_sqr_err, 'MSE': mse, 'MSQRT': msqrt}
    tmp = pd.DataFrame(new_row,index=[0])
    df_err = pd.concat([df_err,tmp], ignore_index=True)

print(df_err.head(10))
```

```python
df_err.sort_values(by='MSE', ascending=True, inplace=True)
print(df_err.head(50))
```

```python
# save to csv file
df_err.to_csv('Mdl_err1.csv')
df_err_100 = df_err.head(100)
df_err_100.to_csv('Mdl_err1_top100.csv')
```
